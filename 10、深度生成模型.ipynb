自编码器：



我们使用VAE来生成新的图像，它其实类型是一种高斯混合分布...

对于图像、文本的生成，我们可以使用PixelRNN，VAE，GAN...

为什么要使用VAE？

如果我们单纯地使用自编码器，我们只能得到指定输入的图像，并不能生成其它的随机的新的图像，如果需要生成任意图像，就需要在隐含向量的基础上加入一个正太分布的噪声，这个时候就不会像自编码器那样只生成一个隐含向量，而是改为生成一个均值向量和一个标准差向量，通过标准差乘以一个标准正太分布再加上均值向量，得到银行向量，作为解码器的输入。

VAE变分自编码器，主要思路是使用神经网络来训练样本的分布，因为我们的训练样本数据默认是不知道其分布的，而又要求得到其分布函数，进而可以生成分布中的其它数据，如果知道这个样本分布的话，也就无需使用神经网络了。所以对于这种问题，我们只能假设推断，虽然不知道，它应该隐含着属于某种分布，引入隐变量，假设隐变量属于正态分布。

我们知道每个序列数据可能会属于某个分布，而对于多个序列它应该是多个随机变量，而不应该把多个变量设定满足一个分布，这是一种对随机变量概率分布的概念误解，所以应该假设每个序列符合一个比如正太分布，这样得到的概率分布而后生成的其对应的数据才是可解释的。比如说我们的每个图片，或每个句子都可以假定为一个分布。（但是，如何某个序列内的数据不够多，是无法使用EM算法的概率来评估，而更应该使用最大后验概率，而不是最大化似然估计，也就只能使用贝叶斯公式来表示）
