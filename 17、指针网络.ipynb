指针网络的概念是从原始数据中选取期望的一部分数据，比如在机器翻译中，对于某些人名、地名是不需要翻译的，希望直接能够从原始数据中截取这部分数据到翻译后的句子中，聊天机器人也是类似的。

我们在使用传统的序列生成时，对于这种出现人名、地名的情况，一般效果不会很好，翻译中语料库中基本不会出现这些人名、地名，使用指针网络就可以实现。

使用序列到序列的方式来输出指定长度的序列，存在的一个问题，就是太死了，不能动态生成。而后就想到了加入注意力机制，这种动态决策的方法。比如输入的是所有的文档的词向量和标记，期望输出动态可变长度的文章中的词，按照这种思路，我们在对文本做总结时，感觉很有用。



指针网络主要还是针对原始数据存在不需要处理的词的解决办法...
