递归网络，其实是对循环网络的一种变形，从横着放，变成竖着放，不再传递隐状态的信息，而是隐状态信息的叠加，如下图所示：



在情感分析中，有这么一个需求，对于句子中褒贬词和强化词，我们希望能够分开，和结合处理，使得句子的语义更清楚，比如not very good，very是强化词，应与much结合处理，而not是否定词，应分开处理。

但是需要注意的是，在对very good处理后的输出的维度需要与not的维度一致，才能再次结合作为输入。对于very good的表达我们不能简单地对两向量相加，就用于表示very good的语义，是需要经过函数映射后才行的，也就是说是需要神经网络来训练的。

对于情感分析的输出类别中，一般会按照层级来定义类别，如特别好，很好，好，不太好，不好，很差等...
