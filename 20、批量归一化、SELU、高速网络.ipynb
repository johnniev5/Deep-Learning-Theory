批量归一化，对应着批量随机梯度，也就是这一个批次的数据进行feature scaling操作，使其满足均值为0，方差为1的标准正态分布，既然是要求满足正太分布，那就要求数据量要足够，太少是不满足使用频率统计的。

在使用批量归一化训练时，是批量进行的，而在测试时是没有进行批量处理的，那该如何做呢？思路就是使用全量数据计算移动平均。

批量归一化的好处：

1. 可以减少训练时间，可以训练更深的网络
2. 可以解决梯度消失和梯度爆炸的问题，比如在使用sigmoid、tanh等激活函数
3. 学习时，受初始化参数的影响更小
4. 减少了对正则化的需求

除了有批量归一化，还有：

* 批量重归一化
* 层归一化
* 实例归一化
* 权重归一化
* 光谱归一化
