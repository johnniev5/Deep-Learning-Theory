如果单纯地使用DNN来通过图像的序列来识别图像的话，有一个很明显的问题，就是随着图片分辨率的增大，其像素的个数也会增大，也就带来特征的增加。且图片中存在好多pattern是不需要输入整副图像的，这也就需要考虑对于图片中的局部同类的识别，只需要一个很小的pattern取匹配就行，这就一定程度上减少了参数量。另外，我们根本就需要原始图像的像素作为特征，即使是少量的模糊特征，也是可以识别的，这就是对图片进行subsampling，就是我们经常说的pooling操作。

一般CNN的组成有：卷积层 + 卷积层 + 池化层（可以自由变换）+ Flatten + 全连接层 + 分类器

CNN中的卷积层一般会有多个卷积核，每个卷积核是对一种模式的识别，比如横纹，竖纹，斜纹等基本的纹理的识别。卷积核的数目一般是自定义的，且不同的卷积核通过训练得到的数据是不一致的。但每个卷积核对于某种pattern的识别是一致的，也就说是同pattern权重参数的。多个卷积核组合一个feature map，也就从图像中固定的特征模式。

卷积核的卷积操作，其实就是卷积核与窗口大小的图片区域的每个元素进行按位相乘再相加得到的标量数据。窗口每次移动的步长和卷积核的大小决定了经卷积操作后的结果矩阵大小。

一般在大像素的图像中，建议卷积核的大小设置的大点，如设置成11, 9等，相应的步长也应该大小，这样子能够减少参数运算量。

对于灰度图像的卷积操作，因其通道维度是1，所以在卷积操作时每次也只进行一次，而彩色图像的颜色通道是3，每次卷积操作会是3次，且这三个通道的卷积矩阵还是一样的。

如果完全使用全连接的NN来训练的话，存在参数巨大的问题，而选择使用CNN，从可以看到卷积核的操作其实就是对局部对象的识别，而既然是局部，固然其使用的参数要大大减少，另外，再使用相同pattern的权重参数是共享的，又再一次减少了参数量。

其实，我们进一步探究可知，卷积和池化操作，就是将一副图像，分成几个pattern来分别进行处理和识别，每个pattern提取固定的模式。这也符合将复杂事情简单化，分步骤处理原则。

在使用Keras进行图像处理时，可以直接将Conv作为输入，需要加入input_shape参数，而如果在此之前有Embedding层，则无需在此加入，对于自定义模型也是如此，需要预先定义Input层。

我们知道对于CNN的应用，它在浅层是对某些纹理的识别，而在深层，则是对某些固定对象的识别，也就是多个纹理的组合形成的形状。这是CNN比较有趣的地方。

CNN的损失函数比较有意思，就是对卷积操作后的输出的所有元素和要求最大输出，或者更准确地来说，是对全连接层之后的输出最大化，也就是输出最大化正确性。

对于Deep Dream这种项目，它就是通过训练CNN来最大化它看到的内容，最终输出的效果会夸大化自己看到的内容。

像Deep Style这样的项目，是让模型看艺术类的图像，通过对艺术类的图像的学习，能够将一些普通的图项加入艺术元素。

而对于AlphaGo，其中也用到了监督式CNN训练和强化学习部分。因为对于棋谱上的方格就类似图像中的像素，同样可以使用CNN来进行训练。AlphaGo中并没有使用到maxpooling层。

同样对于语音的识别，也可以使用CNN来训练，输入原始语音数据；一维卷积可以用来对文本的训练...
