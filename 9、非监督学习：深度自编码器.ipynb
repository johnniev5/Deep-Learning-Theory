非监督学习更加符合人类的学习方式，所以说非监督学习才是机器学习新的未来。

自编码是一种最小化重构误差的方法，一是可以用它来降维，而是用其进行编码，比如词向量的低维表示。

我们自此可以联想到PCA降维，因PCA降维它是一种有损失的降维方法，相比于使用自编码器，它是无损的，由此我们可以将其应用于对图像的大小修改，音频压缩等。

采用深度自编码进行降维，最终可视化的数据分得更开，且我们在原始数据中加入噪声，自编码器依然能够很好地还原原数据。

对于非线性降维，可以使用tSNE，自编码器，而线性降维一般选择PCA。

我们在使用词袋模型表示词向量的时候，是没有考虑上下文语义的，比如使用one-hot编码，各个词向量间在词向量空间中都是互相垂直的，也就无法计算相关度。

而使用自编码就不一样，它是将高维稀疏向量降维为低维稠密向量，将高维坐标轴上的数据映射到向量空间中，这就使得向量间有了可比较性，同时在可视化上使用自编码的降维效果要好于一般方法。

使用自编码器来查找相似的图片，比如我们将不同分辨率的图像拉平成向量，然后使用自编码器降维到同纬度，再使用余弦相似度找到相似的TOPN的图片。

在训练自编码时，我们需要能够保证原数据的尽量还原，得到模型后，可以取前半部分自编码为应用。

我们在CNN中的卷积层和池化层的功能和自编码器的编码阶段很像，由此我们可以想到如何应用自编码的解码阶段，就是Deconvolution，来对图像进行增色，一般的Deconvolution有使用相关部位数值填充法。
